```python
import numpy as np
import pandas as pd
import keras
import sklearn

from sklearn.preprocessing import OneHotEncoder, OrdinalEncoder
from sklearn.pipeline import Pipeline

"""
- Da es sich um ein Regressionsproblem handelt, hat der Outputlayer keine Aktivierungsfunktion
- Der Optimizer Adam wird verwendet als wie im vorigen Bsp. bei MNIST Stochastic Gradient Descent SGD
- Loss Funktion wird mit means square error berechnet
"""
```

```python
"Laden der Daten"

data = 'fake_job_postings.csv'

df = pd.read_csv(data)

df.head()
```

```python
df_neu = pd.get_dummies(df)
df_neu
```

```python
"Erstellung von Trainingsdaten und Testdaten"

from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(df_neu, df_neu['fraudulent'], test_size=0.2, random_state=42, stratify=df_neu['fraudulent'])
```

```python
x_test.shape
```

```python
"Standardscaler verwenden"

from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
x_train = sc.fit_transform(x_train)
x_test = sc.transform(x_test)
```

```python
import keras
from keras.layers import Dense, Activation,Dropout
from keras.models import Sequential

model = Sequential()

model.add(Dense(128,activation  = 'relu',input_dim =x_test.shape[1]))
model.add(Dense(64,activation  = 'relu'))
model.add(Dense(32,activation  = 'relu'))
model.add(Dense(16,activation  = 'relu'))
model.add(Dense(1))
model.compile(optimizer = 'adam',loss = 'mean_squared_error')

"""
Veränderung der Layergrösse --> siehe MNIST
"""
```

```python
model.fit(x_train, y_train, epochs = 10)
```

```python
#Evaluierung

y_pred = model.predict(x_test)
```

```python
"Genauigkeit"
from sklearn.metrics import r2_score
r2 = r2_score(y_test, y_pred)
print(r2)
```

```python
"RMSE"

from sklearn.metrics import mean_squared_error
rmse = (np.sqrt(mean_squared_error(y_test, y_pred)))
print(rmse)
```
